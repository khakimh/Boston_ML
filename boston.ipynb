#### Import python package

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston 
# from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

### Load Boston house prices dataset and creating DataFrame
###### # creating DataFrame from data 2D (506,13) and the columns name was separated
###### # add new columns by df['columns_name'] = data_1D

```python
boston = load_boston()
df = pd.DataFrame(
    boston['data'],
    columns=boston['feature_names']
)
df['PRICE'] = boston['target']
df
```

#### Checking data's correlation

```python
korelasi = df.corr()
korelasi
```

#### Creating correlations bar

```python
plt.imshow(korelasi, cmap='BuPu')
plt.colorbar()
plt.yticks(np.arange(14), df.columns)
plt.xticks(np.arange(14), df.columns)
plt.xticks(rotation=90)
plt.show()
# np.arange(0,14)
```

#### Machine Learning Model

```python
model = linear_model.LinearRegression()
model.fit(df[['ZN', 'CHAS', 'RM', 'DIS', 'B']], df['PRICE'])
```

#### Predicting

```python
# df[['ZN', 'CHAS', 'RM', 'DIS', 'B']].iloc[0] #not working because its 1D
df.head(1)[['ZN', 'CHAS', 'RM', 'DIS', 'B']] #working because its 2D
print('house price base on predicting = ', round(model.predict(df.head(1)[['ZN', 'CHAS', 'RM', 'DIS', 'B']])[0],2))
print('house price base on data = ', list(df.head(1)['PRICE'])[0])
```
